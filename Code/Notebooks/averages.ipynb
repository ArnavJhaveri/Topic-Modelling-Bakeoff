```python
# Trying to use PCA

import pickle
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

dataset = 'twitter'
model = 'BERTopic'

train = pickle.load(open(dataset + '/' + model + '/run_1/train.pkl', 'rb'))
test = pickle.load(open(dataset + '/' + model + '/run_1/test.pkl', 'rb'))
temp = pd.concat([train, test]).reset_index(drop=True) # concatenating train and test datasets
df = pd.read_csv(dataset + '.csv') # loading in actual dataset

for run in range(1, 2):

    test_distribution = pickle.load(open(dataset + '/' + model + '/run_' + str(run) + '/' + model + '_topic_distribution_test.pkl', 'rb'))
    train_distribution = pickle.load(open(dataset + '/' + model + '/run_' + str(run) + '/' + model + '_topic_distribution_train.pkl', 'rb'))
    distribution = np.concatenate([train_distribution, test_distribution]) # concatenating train and test distributions

    merged = pd.merge(temp, df, how='inner', left_on = 'message_id', right_on = 'Unnamed: 0')[['message_id_x', 'message_x', 'user_id']]
    distribution_df = pd.concat([pd.DataFrame(distribution), merged[['user_id']]], axis = 1)

    # mean
    user_distributions = distribution_df.groupby('user_id').mean().reset_index(drop=True)
    user_distributions_arr = [row[1].values for row in user_distributions.iterrows()]

    # PCA then mean
    scaler = StandardScaler()
    standardized_data = scaler.fit_transform(distribution_df.drop('user_id', axis=1))
    
    # Apply PCA
    pca = PCA(n_components=2)  # Reduce to 2 components for demonstration
    reduced_data = pca.fit_transform(standardized_data)

    # Create a DataFrame for the reduced data
    reduced_df = pd.DataFrame(reduced_data, columns=['PC1', 'PC2'])
    reduced_df['user_id'] = merged['user_id']

    # Aggregate these PCA components by user
    pca_user_distributions = reduced_df.groupby('user_id').mean().reset_index(drop=True)
    pca_user_distributions_arr = [row[1].values for row in pca_user_distributions.iterrows()]

    # with open(dataset + "/" + model + "/run_" + str(run) + "/" + model + '_avg_topic_distribution.pkl', 'wb') as f:
    #     pickle.dump(user_distributions_arr, f)
    
    print("Done " + str(run))
```

```python
import pandas as pd
import pickle
import csv
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import r2_score

# change this
dataset = 'twitter' # folder and dataset name
model = "CTM" # model name
cols = ['gender', 'age', 'politics'] # outcome columns

for run in range(1, 6):
    print("Run: " + str(run))
    print()

    # loading in distributions that were saved during topic extraction
    test_distribution = pickle.load(open(dataset + '/' + model + '/run_' + str(run) + '/' + model + '_topic_distribution_test.pkl', 'rb'))
    train_distribution = pickle.load(open(dataset + '/' + model + '/run_' + str(run) + '/' + model + '_topic_distribution_train.pkl', 'rb'))
    train = pickle.load(open(dataset + '/' + model + '/run_' + str(run) + '/train.pkl', 'rb'))
    test = pickle.load(open(dataset + '/' + model + '/run_' + str(run) + '/test.pkl', 'rb'))

    topics = []
    with open(dataset + '/' + model + '/run_' + str(run) + '/' + 'topics_100.txt', 'r') as f:
        reader = csv.reader(f)
        for row in reader:
            topic_list = [item.strip() for item in row if item.strip()]
            topics.append(topic_list)

    distribution = np.concatenate([train_distribution, test_distribution]) # concatenating train and test distributions
    temp = pd.concat([train, test]).reset_index(drop=True) # concatenating train and test datasets
    df = pd.read_csv(dataset + '.csv') # loading in actual dataset

    df['politics'] = round(df['politics']) # for twitter

    merged = pd.merge(temp, df, how='inner', left_on = 'message_id', right_on = 'Unnamed: 0')[['message_id_x', 'user_id', 'message_x'] + cols]
    merged.columns = ['message_id', 'user_id', 'message'] + cols

    user_ids = {user_id : user_data.index.tolist() for user_id, user_data in merged.groupby('user_id')} # gets all users and their corresponding indices in the df
    all_user_ids = list(user_ids.keys())

    for outcome in cols:

        labels_data = merged[['user_id'] + cols] # df of user_ids and all the labels to consider
        user_ids_avg_dict = {} # populated to be user_id : average distribution
        user_ids_avg = [] # populated to be all the average distributions ^ same as the values from there
        y = [] # the outcome we care about

        for key, value in user_ids.items():
            arrays = [distribution[i] for i in value]
            avg = np.mean(arrays, axis = 0)
            user_ids_avg_dict[key] = avg
            user_ids_avg.append(avg)
            y.append(labels_data[labels_data['user_id'] == key][outcome].iloc[0])

        X = user_ids_avg

        # 80-20 split --> didn't use train-test-split function since its already shuffled
        X_train = X[:round(0.80 * len(X))]
        X_test = X[round(0.80 * len(X)):]

        # bucketing twitter ages -- did it by 10 but we can make this so that each bucket has the same number of users if we'd like
        if outcome == "age":
            n = []
            for i in y:
                if i < 10:
                    n.append("age_0to10")
                elif i < 20:
                    n.append("age_10to20")
                elif i < 30:
                    n.append("age_20to30")
                elif i < 40:
                    n.append("age_30to40")
                elif i < 50:
                    n.append("age_40to50")
                elif i < 60:
                    n.append("age_50to60")
                elif i < 70:
                    n.append("age_60to70")
                elif i < 80:
                    n.append("age_70to80")
                elif i < 90:
                    n.append("age_80to90")
                else:
                    n.append("age_90to100")
            y = n

        y_train = y[:round(0.80 * len(X))]
        y_test = y[round(0.80 * len(X)):]

        train_users = all_user_ids[:round(0.80 * len(X))]
        test_users = all_user_ids[round(0.80 * len(X)):]

        # converts all outcomes into dummy variables
        dummy_train = pd.get_dummies(y_train, dtype=int)
        dummy_test = pd.get_dummies(y_test, dtype=int)
        dummy_train.columns = ["label " + str(i) for i in range(len(dummy_train.columns))]
        dummy_test.columns = ["label " + str(i) for i in range(len(dummy_test.columns))]

        accuracies = []
        r2s = []
        for i, label_col in enumerate(list(dummy_train)):
            lr_model = LogisticRegression().fit(X_train, dummy_train[label_col]) # trains logit model over each label column from dummy df
            accuracies.append(lr_model.score(X_test, dummy_test[label_col])) # gets scores on test set
            r2s.append(r2_score(dummy_test[label_col], lr_model.predict(X_test))) # gets r2 score from test set

        accuracies = [np.mean(accuracies)] + accuracies + [np.mean(r2s)] # appends to list for final_df below
        final_df = pd.concat([pd.Series(accuracies)], axis = 1).T
        final_df.columns = ['Average Accuracy'] + [i + str(" Accuracy") for i in dummy_train.columns] + ['R2']
        print(outcome)
        display(final_df)
        print()
        print()
```
